**BAR GRAPHS**

```python
ss_df=pd.DataFrame({"season":["spring","summer","fall","winter"],"No.of.students":[5,12,6,10]})
```

```python
plt.figure(figsize=(8,6))
plt.bar(x=ss_df['season'], 
        height =ss_df['No.of.students'],
color=["red","green","blue","purple"])
plt.title('Arrival Of Students By Season')
plt.xlabel('Seasons')
plt.ylabel('No of Students')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot((data=ss_df,x='season'), 
        y =('No.of.students'),
polette=("red","green","blue","purple"))
plt.show()
```

**Piecharts**

```python
x=np.array([50,90,100,15,10])
labels=np.array(['SEE','+2','BACHELOR','MASTER','PHD'])
```

```python
plt.figure(figsize=(8,6))
plt.pie(x,labels=labels,autopct ='%.1f%%',explode=[0,0.03,0,0,0.2])
plt.legend()
plt.title('students ledger')
plt.show
```
